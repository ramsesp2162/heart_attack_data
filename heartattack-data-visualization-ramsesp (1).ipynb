```python
'''
Author: Ramses Partida
Date: 04/24/2021
ISTA 131 
'''
import numpy as np
import pandas as pd 
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\ramse\Documents\ISTA131\final\heart.csv')
df.head()
```

```python
def barplot(data, x, y, hue):
    fig, ax = plt.subplots(1, 1, figsize=(15, 10))
    sns.barplot(data=data, x=x, y=y, hue=hue)
    plt.show()
```

```python

df['sex'] = np.where(df['sex']==1, 'male', 'female')
df['cp'] = np.where(df['cp']==0, 'typical angina', np.where(df['cp']==1, 'atypical angina', np.where(df['cp']==2, 'non-anginal pain', np.where(df['cp']==3, 'asymptomatic', 0))))
```

```python
# thalachh = max heart rate
# comparative barplot showing the types of chest pain(cp) distributed among sex and max heart rate
barplot(data = df, x='sex', y='thalachh', hue='cp')
```

```python
#heat map showing the correlation between variables, 
# results indicate no redundancies
X = df.drop('output', axis=1)
y = df['output']

fig, ax = plt.subplots(1, 1, figsize=(15, 10))
sns.heatmap(X.corr(), annot=True)
plt.show()
```
